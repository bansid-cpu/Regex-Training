```python
plt.figure(figsize=(12,8))
sns.barplot(x = 'Total Parking Spaces',y = 'Property Type',data = parking,linewidth = 1.5)
plt.title('Total Parking Spaces Vs Property Type')
```

```python
plt.figure(figsize =(14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = 'Bldg Status',y = 'Total Parking Spaces',hue = 'Property Type',markers = True,data = parking,linewidth = 2)
```

```python
plt.figure(figsize=(14,7))
sns.set(rc={'xtick.labelsize':20,'ytick.labelsize':15,'axes.labelsize':15 , "axes.grid":False})
sns.lineplot(x="Owned/Leased" , y="Total Parking Spaces" , hue = "Property Type" ,style="Property Type" ,dashes = False,palette = 'CMRmap',data=parking)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.gcf().text(.5,.9,'Scatter Plot',fontsize = 40,color = 'Black',ha = 'center',va = 'center')
sns.scatterplot(x = 'Bldg Status',y = 'Total Parking Spaces',hue = 'Owned/Leased',data = parking)
```

## 1. What is values for owned/leased

```python
owned_leased = pd.pivot_table(parking,index = 'Owned/Leased',values = 'Total Parking Spaces')
```

```python
owned_leased
```

```python
plt.bar(owned_leased.index,owned_leased['Total Parking Spaces'])
```

## 2. In which bldg state parking situation is in excess 

```python
excess_parking = pd.pivot_table(parking,index = 'Bldg State',columns = 'Bldg Status',values='Total Parking Spaces')
```

```python
excess_parking
```

```python
excess_parking = excess_parking[excess_parking.EXCESS>0]
```

```python
excess_parking
```

```python
plt.bar(excess_parking.index,excess_parking.EXCESS)
```

```python
excess_parking_sorted = excess_parking.EXCESS.sort_values()
```

```python
excess_parking_sorted
```

```python
plt.bar(excess_parking_sorted.index,excess_parking_sorted.values)
```
